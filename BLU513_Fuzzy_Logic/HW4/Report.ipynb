```python
import matplotlib.pyplot as plt
import numpy as np
import skfuzzy

plt.rc('text', usetex=True)
plt.rc('text.latex', preamble=r'\usepackage{amsmath}')
plt.rc('font', family='sans-serif')
```

```python
X = np.linspace(0,10,101)
A1 = skfuzzy.trimf(X, (0,2,6))
A2 = skfuzzy.trimf(X, (1,6,10))

plt.plot(X, A1, label='$A_1$')
plt.plot(X, A2, label='$A_2$')
plt.text(11, 0.5, 
        r'$ \mu(z) = \begin{cases}'
        r'    \frac{z}{2} & z \in [0,2] \\'
        r'    \frac{6-z}{4} & z \in [2,\frac{34}{9}] \\'
        r'    \frac{z-1}{5} & z \in [\frac{34}{9},6] \\'
        r'    \frac{10-z}{4} & z \in [6,10] \\'
        r'\end{cases} $', {'fontsize': 20}, horizontalalignment='left')

plt.axvline(4, label='MoM', ls='dashdot', c='grey')
plt.axvline(2, label='FoM', ls='dashed', c='grey')
plt.axvline(6, label='LoM', ls='dotted', c='grey')

plt.axvline(4.6653, label='CoG', ls='dashed', c='#66BB66')
plt.axvline(4.4667, label='CoGoA', ls='dashed', c='#BB6666')
plt.axvline(6.2492, label='CoLA', ls='dashed', c='#6666BB')

plt.legend()
plt.show()
```

1.  Max membership principle:<br>
    $z^* = \{z | \mu(z) \gt \mu(x), \text{for all }x\}$<br>
    Since we have two maxima, we can't apply it.

2.  Middle of maxima (MoM):<br>
    $z^* = \frac{\sum{\overline{z}}}{n}$, where $\overline{z} = \{z | \mu(z) \ge \mu(x), \text{for all }x\}$ , $n = |\overline{z}|$<br>
    $z^* = \frac{2+6}{2} = 4$

3.  First/Last of maxima (LoM/FoM):<br>
    $z^*$ is the respectively the minimum and maximum value of $\overline{z}$ (See 2.)<br>
    $z^* = 2 | 6$

4.  Weighted average method:<br>
    $z^* = \frac{\sum{\mu_C(\overline{z}) \cdot \overline{z}}}{\sum{\mu_C(\overline{z})}}$, which can be expressed as the sums of the MoM weighted by their membership values, divided by the sum of the membership value of the maximas. The sum is apply on each component of the fuzzy set.<br>
    However, each component need to be symmetric, which is not our case.

5.  Centroid method (Center of gravity, CoG):<br>
    $ z^* = \frac{\int{\mu(z) \cdot z\ dz}}{\int{\mu(z)\ dz}}$<br>
    $z^* = \frac{\int_0^2 z \cdot 0.5z + \int_2^{34/9} z \cdot (1.5-0.25z) + \int_{34/9}^6 z \cdot (0.2z - 0.2) + \int_6^{10} z \cdot (2.5-0.25z) }{ \int_0^2 0.5z + \int_2^{34/9} 1.5-0.25z + \int_{34/9}^6 0.2z - 0.2 + \int_6^{10} 2.5-0.25z } = 4.6653$

6.  Center of sums (center of gravity of ares):<br>
    The same as the centroid, but using the initials fuzzy sets instead of their unions<br>
    $z^* = \frac{\int_0^2 z \cdot 0.5z + \int_2^6 z \cdot (1.5-0.25z) + \int_1^6 z \cdot (0.2z - 0.2) + \int_6^{10} z \cdot (2.5-0.25z) }{ \int_0^2 0.5z + \int_2^6 1.5-0.25z + \int_1^6 0.2z - 0.2 + \int_6^{10} 2.5-0.25z } = 4.4667$

7.  Center of largest area:<br>
    The same as the centroid of the largest convex part of the set.<br>
    $ A_1 = \int_0^2 0.5z + \int_2^{34/9} 1.5-0.25z = 2.3827 $<br>
    $ A_2 = \int_{34/9}^6 0.2z - 0.2 + \int_6^{10} 2.5-0.25z = 3.7284 $<br>
    $ z^* = \frac{ \int_{34/9}^6 z \cdot (0.2z - 0.2) + \int_6^{10} z \cdot (2.5-0.25z) }{ 3.7284 } = 6.2492 $<br>
